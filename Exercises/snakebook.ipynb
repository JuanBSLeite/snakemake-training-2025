## Basic Snakemake workflow

Write the python script src/Basic/plotHistogram.py that receives as args input and output files and plot the dalitz plot.

```python
os.makedirs("src/Basic", exist_ok=True)
os.makedirs("data", exist_ok=True)
os.makedirs("plots/basic", exist_ok=True)

with open("src/Basic/Snakefile", "w") as f:
    f.write(
        """
rule plotHistogram:
    input:
        root_file="data/BuToPiPiPi.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/basic/Dalitz.png"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/Basic/Snakefile --cores 1")
```

## Environment setup with Conda

Let's first install miniconda.

```python
os.system("source installMiniconda.sh")
```

```python
# add miniconda to PATH

import sys
conda_bin = os.path.expanduser("miniconda3/bin")


# Make conda executable visible
os.environ["PATH"] = f"{conda_bin}:{os.environ['PATH']}"

# Where snakemake stores environments
os.environ["SNAKEMAKE_CONDA_PREFIX"] = os.path.expanduser(
    ".snakemake/conda"
)
```

```python
# let's create the yaml file for the environment

os.makedirs("src/Conda", exist_ok=True)
os.makedirs("plots/conda", exist_ok=True)

with open("src/Conda/env.yaml", "w") as f:
    f.write(
        """
name: my_environment
channels:
  - defaults
  - conda-forge
dependencies:
  - python
  - numpy
  - matplotlib
  - pandas
  - uproot
        """
    )
```

```python
os.system("conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/main")
os.system("conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/r ")
```

```python
with open("src/Conda/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]
ext = ["png", "pdf"]

rule all:
    input:
        expand("plots/conda/{sample}.{ext}", sample=samples, ext=ext)

rule plotHistogram:
    input:
        "data/{sample}.root",
    output:
        "plots/conda/{sample}.{ext}"
    conda:
        "env.yaml"
    script:
       "plotHistogram.py"
       """
    )
       
os.system("snakemake --use-conda -s src/Conda/Snakefile --cores 1")
```
