```python
import os
```

## Basic Snakemake workflow

```python
os.makedirs("src/Basic", exist_ok=True)
os.makedirs("data", exist_ok=True)
os.makedirs("plots/basic", exist_ok=True)

with open("src/Basic/Snakefile", "w") as f:
    f.write(
        """
rule plotHistogram:
    input:
        root_file="data/BuToPiPiPi.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/basic/Dalitz.png"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/Basic/Snakefile --cores 1")
```

## Generalizing rules

```python
os.makedirs("src/GeneralizingRules", exist_ok=True)
os.makedirs("plots/generalizing_rules", exist_ok=True)

with open("src/GeneralizingRules/Snakefile", "w") as f:
    f.write(
        """
rule plotHistogram:
    input:
        root_file="data/{sample}.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/generalizing_rules/{sample}.png"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/GeneralizingRules/Snakefile plots/generalizing_rules/BuToPiPiPi.png --cores 1")
os.system("snakemake -s src/GeneralizingRules/Snakefile plots/generalizing_rules/BuToKKK.png --cores 1")
```

## Aggregating Inputs

```python
os.makedirs("src/AggregatingInputs", exist_ok=True)
os.makedirs("plots/aggregating_inputs", exist_ok=True)

with open("src/AggregatingInputs/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]

rule all:
    input:
        expand("plots/aggregating_inputs/{sample}.png", sample=samples)

rule plotHistogram:
    input:
        root_file="data/{sample}.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/aggregating_inputs/{sample}.png"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/AggregatingInputs/Snakefile --cores 1")
```

## Aggregating Inputs II

```python
os.makedirs("src/AggregatingInputsII", exist_ok=True)
os.makedirs("plots/aggregating_inputs_ii", exist_ok=True)

with open("src/AggregatingInputsII/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]
ext = ["png", "pdf"]

rule all:
    input:
        expand("plots/aggregating_inputs_ii/{sample}.{ext}", sample=samples, ext=ext)

rule plotHistogram:
    input:
        root_file="data/{sample}.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/aggregating_inputs_ii/{sample}.{ext}"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/AggregatingInputsII/Snakefile --cores 1")
```

## Aggregating Inputs III

```python
os.makedirs("src/AggregatingInputsIII", exist_ok=True)
os.makedirs("plots/aggregating_inputs_iii", exist_ok=True)

with open("src/AggregatingInputsIII/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]
ext = ["png", "pdf"]

outputs = []

for sample in samples:
    for e in ext:
        outputs.append(f"plots/aggregating_inputs_iii/{sample}.{e}")

rule all:
    input:
        outputs

rule plotHistogram:
    input:
        root_file="data/{sample}.root",
        script="src/Basic/plotHistogram.py"
    output:
        "plots/aggregating_inputs_iii/{sample}.{ext}"
    shell:
        "python {input.script} {input.root_file} {output}"
        """
    )
       
os.system("snakemake -s src/AggregatingInputsIII/Snakefile --cores 1")
```

## Using custom scripts

```python
os.makedirs("src/CustomScripts", exist_ok=True)
os.makedirs("plots/custom_scripts", exist_ok=True)

with open("src/CustomScripts/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]
ext = ["png", "pdf"]

rule all:
    input:
        expand("plots/custom_scripts/{sample}.{ext}", sample=samples, ext=ext)

rule plotHistogram:
    input:
        "data/{sample}.root",
    output:
        "plots/custom_scripts/{sample}.{ext}"
    script:
       "plotHistogram.py"
       """
    )
       
os.system("snakemake -s src/CustomScripts/Snakefile --cores 1")
```

## Environment setup with Conda

Let's first install miniconda.

```python
os.system("source src/installMiniconda.sh")
```

```python
# add miniconda to PATH

import sys
conda_bin = os.path.expanduser("miniconda3/bin")


# Make conda executable visible
os.environ["PATH"] = f"{conda_bin}:{os.environ['PATH']}"

# Where snakemake stores environments
os.environ["SNAKEMAKE_CONDA_PREFIX"] = os.path.expanduser(
    ".snakemake/conda"
)
```

```python
# let's create the yaml file for the environment

os.makedirs("src/Conda", exist_ok=True)
os.makedirs("plots/conda", exist_ok=True)

with open("src/Conda/env.yaml", "w") as f:
    f.write(
        """
name: my_environment
channels:
  - defaults
  - conda-forge
dependencies:
  - python
  - numpy
  - matplotlib
  - pandas
  - uproot
        """
    )
```

```python
os.system("conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/main")
os.system("conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/r ")
```

```python
with open("src/Conda/Snakefile", "w") as f:
    f.write(
        """
samples = ["BuToPiPiPi", "BuToKKK"]
ext = ["png", "pdf"]

rule all:
    input:
        expand("plots/conda/{sample}.{ext}", sample=samples, ext=ext)

rule plotHistogram:
    input:
        "data/{sample}.root",
    output:
        "plots/conda/{sample}.{ext}"
    conda:
        "env.yaml"
    script:
       "plotHistogram.py"
       """
    )
       
os.system("snakemake --use-conda -s src/Conda/Snakefile --cores 1")
```
